### Pixel classification

For this plugin, you will rely on the `scikit-learn` package which ships various classification algorithms. Your plugin should enable users to select between three different classification algorithms: `RandomForestClassifier`, `KNeighborsClassifier` and `MLPClassifier`. Users should also be able to fine-tune key settings of each algorithm. 

To achieve robust classification performance, it is advisable to provide the classification algorithms with feature vectors corresponding to each of the pixels. Here, you should use the `scikit-image` package. 

When training your classifier, you should keep in mind that the data will have to be reshaped.

- [Documentation for `skimage.filters`](https://scikit-image.org/docs/stable/api/skimage.filters.html)

### Random Forest
Users should be able to modify two parameters of the random forest classifier: `n_estimators` and `max_depth`.

- [Documentation for `sklearn.ensemble.RandomForestClassifier`](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html)

### K Nearest Neighbour
Users should be able to modify two parameters of the k nearest neighbour classifier: `n_neighbors` and `metric`.

- [Documentation for `sklearn.neighbors.KNeighborsClassifier`](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html)

### Multilayer Perceptron
Users should be able to modify three parameters of the multilayer perceptron: `hidden_layer_sizes`, `activation` and `solver`. To set `hidden_layer_sizes` use a `TupleEdit` widget.

- [Documentation for `sklearn.neural_network.MLPClassifier`](https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html)
- [Documentation for `TupeEdit`](https://pyapp-kit.github.io/magicgui/api/widgets/TupleEdit/)
